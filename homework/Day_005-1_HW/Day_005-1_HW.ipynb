```python
import pandas as pd
import numpy as np
```

```python
def max_count(data):
    rank_data = data.groupby(by="人口")['國家'].max()
    temp = rank_data.iloc[0]
    max_value = data['人口'].max()
    print (temp+':'+max_value)
```

```python
data = {'國家': ['Tawain','USA','china','japan'],
        '人口': ['23,592,598','329,470,000','1,395,380,000','126,230,000']}
data = pd.DataFrame(data)
```

```python
max_count(data)
```
