## 練習時間
假設我們資料中類別的數量並不均衡，在評估準確率時可能會有所偏頗，試著切分出 y_test 中，0 類別與 1 類別的數量是一樣的 (亦即 y_test 的類別是均衡的)

```python
import numpy as np
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200)
y[:40] = 1
```

```python
y
```

可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
from sklearn.model_selection import train_test_split
```

```python
X1_train ,X1_test , y1_train , y1_test  = train_test_split (X[:40], y[:40], test_size = (1/4))
#   40X = 10, X = 1/ 4 = 0.25
X2_train ,X2_test , y2_train , y2_test  = train_test_split (X[40:], y[40:], test_size = (1/16))
# 160X = 10, X = 1/16 = 0.0625
ytest = np.concatenate ((y1_test, y2_test))
```

```python
X_train ,X_test , y_train , y_test = train_test_split (X, y, test_size = 10,)
```
