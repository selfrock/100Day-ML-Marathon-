```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
wine = datasets.load_wine()
```

```python
x , x_test , y , y_test = train_test_split(wine.data , wine.target , test_size = 0.2 , random_state = 3)
```

```python
logistic = linear_model.LogisticRegression()

logistic.fit (x,y)
```

```python
predict_result = logistic.predict (x_test)

acc = accuracy_score (y_test , predict_result) 
print (acc)
```

```python
boston = datasets.load_boston()
```

```python
X = boston.data[:,np.newaxis , 2]
```

```python
x , x_test , y , y_test = train_test_split(boston.data , boston.target , test_size = 0.2 , random_state = 3)
```
